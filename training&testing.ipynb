```python
#ML Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#accuracy

from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(pred,y_test)
r2=r2_score(pred,y_test)

print('mean square Error :',mse )
print('R2 score: ',r2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: hours studied vs marks scored
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Features (independent variable)
y = np.array([10, 20, 30, 40, 50])            # Target (dependent variable)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions for the training data and a new input
y_pred = model.predict(X)
new_input = np.array([[6]])
new_pred = model.predict(new_input)

print(f"Predicted marks for 6 hours studied: {new_pred[0]:.2f}")
print(f"Coefficient (slope): {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(new_input, new_pred, color='green', marker='x', s=100, label='Prediction (6 hours)')

plt.title('Linear Regression Example: Hours Studied vs Marks Scored')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.legend()
plt.show()
```

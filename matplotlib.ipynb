```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.plot(x,y,color='k')
plt.title('line plot')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.plot(x,y,'*-.',linewidth=1,markersize=10,color='b')
plt.title('line plot')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()

```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.plot(x,y,'bh--',linewidth=2,markersize=30)
plt.title('line plot')

plt.xlabel('X-axis',color='m')
plt.ylabel('Y-axis',color='m')
plt.grid()
plt.show()
```

```python
#bar graph
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.bar(x,y,label='Boys',width=2,color='c')
plt.title('bar-plot')

plt.xlabel('X-axis',color='m')
plt.ylabel('Y-axis',color='m')
plt.legend()
plt.show()
```

```python
x=[9,3,5,7,2]
y=[2,4,6,8,10]
plt.bar(x,y,label='bar',color='orange')
plt.plot(x,y,label='line')
plt.title('bar & line graph')
plt.xlabel('X-axis',color='m')
plt.ylabel('Y-axis',color='m')
plt.annotate("value",xy=(1,2.5),size=10,color='g')
plt.legend()
plt.show()
```

```python
#histogram
ages=[20,19,13,12,16,18,22,23,29,40,36,45,48]
bins=[10,20,30,40,50]  #btw 10 -20 how many ages are there 
                       #btw 20 -29 how many ages are there 

plt.hist(ages,bins,histtype='bar',width=2)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()


```

```python
#pie chart
slice=[8,10,4,3,5]
activities=['sleeping','phone','study','eat','play']
col=['gold','b','magenta','red','g']
plt.pie(slice,labels=activities,colors=col,startangle=90,
        explode=(0,0,0,1.3,0),shadow=True,autopct='%1.3f%%')

plt.title('pie Chart')
plt.show()
```

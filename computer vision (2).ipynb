```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
#RGB to GRAY
gray=cv2.cvtColor(rgb_img,cv2.COLOR_RGB2GRAY)
cv2.imshow('Gray',gray)
cv2.waitkey(0)
cv2.destroyAllWindow()
```

```python
blur=cv2.GaussianBlur(rgb_img,(15,15),2)
plt.imshow(blur)
plt.show()
```

```python
#edge detection
edge=cv2.Canny(rgb_img,100,700)
plt.imshow(edge,cmap='gray')
plt.show()
```

```python
#capturing image through webcam

cap=cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
else:
    ret:False
    
image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("AIML")
plt.axis('off')
plt.show()
cap.release()
```

```python
vdo = r"D:\cinametic vdo\Sequence 01.mp4"
cv2.namedWindow(vdo)
cap = cv2.VideoCapture(vdo)
if cap.isOpened():
    ret,frame = cap.read()
    print(ret)
    print(frame)
else:
    
    ret = False

while ret:
    ret,frame = cap.read()
    if not ret:
        break
    cv2.imshow(vdo,frame)
    if cv2.waitKey(10)==ord('q'):
            break
cv2.destroyAllWindows()
cap.release()
```

```python
#face detection using Haarcasscade
face=(r"C:\Users\Lenovo\Desktop\july aiml\haarcascade_frontalface_default.xml")
model=cv2.CascadeClassifier(face)

img=cv2.imread(r"C:\Users\Lenovo\Pictures\Camera Roll\ai.JPG")
facepoints=model.detectMultiScale(img)
for x,y,w,h in facepoints:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)

plt.imshow(img[::,::,::-1])
plt.show()
```

```python
#face detection using webcam
face = r"C:\Users\Lenovo\Desktop\july aiml\haarcascade_frontalface_default.xml"
cap = cv2.VideoCapture(0)

ret,frame = cap.read()

while ret:
    ret,frame = cap.read()
    print(ret)
    print(frame)

    if not ret:
        break
    model = cv2.CascadeClassifier(face)
    facepoints = model.detectMultiScale(frame)

    for x,y,w,h in facepoints:
        cv2.rectangle(frame,(x,y), (x+w,y+h), (0,255,0), 3)

    cv2.imshow('Apna Face',frame)
    if cv2.waitKey(10)==ord('q'):
        break
cv2.destroyAllWindows()
cap.release()
```
